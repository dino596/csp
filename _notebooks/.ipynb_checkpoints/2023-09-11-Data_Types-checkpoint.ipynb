#Variables and Data Types

```python
import random # used for random.shuffle

InfoDb = []

InfoDb.append({
    "LastName": "Mansour",
    "Period": 1,
    "Class": "Ethnic Literature I"
})

InfoDb.append({
    "LastName": "Mortensen",
    "Period": 2,
    "Class": "AP Computer Science Principles"
})

InfoDb.append({
    "LastName": "Swanson",
    "Period": 3,
    "Class": "AP United States History"
})

InfoDb.append({
    "LastName": "Ozuna",
    "Period": 4,
    "Class": "AP Chemistry"
})

InfoDb.append({
    "LastName": "Bernabeo",
    "Period": 5,
    "Class": "AP Calculus BC"
})

print("Formatted Print")
for i in range(len(InfoDb)):
    print("Index:", i)
    print("\tTeacher:", InfoDb[i]["LastName"])
    print("\tPeriod:", InfoDb[i]["Period"])
    print("\tClass:", InfoDb[i]["Class"])
    
print("\nReverse Order Print")
for i in range(len(InfoDb)):
    index = len(InfoDb) - 1 - i
    print("Index:", index)
    print("\tTeacher:", InfoDb[index]["LastName"])
    print("\tPeriod:", InfoDb[index]["Period"])
    print("\tClass:", InfoDb[index]["Class"])

print("\nRandomized Order Print") 
random.shuffle(InfoDb)
for i in range(len(InfoDb)):
    print("Index:", i)
    print("\tTeacher:", InfoDb[i]["LastName"])
    print("\tPeriod:", InfoDb[i]["Period"])
    print("\tClass:", InfoDb[i]["Class"])
```
